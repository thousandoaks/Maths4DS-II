# Example 1. Log Hourly Wage, marital status and gender

## In this example we want to test for differences in return to education accross gender.

### Using the data in WAGE1 where n=526 individuals

```python
Wages = woo.dataWoo('wage1')
```

## Let's interact female with educ:

```python
# We impose a simple, linear, model:
# We specify CeoSalaries as the empirical dataset

reg = smf.ols(formula='np.log(wage) ~female+educ+female*educ+exper+np.power(exper,2)+tenure+np.power(tenure,2)', data=Wages)
```

```python
# We fit the model
results = reg.fit()
```

```python
b = results.params
print(f'b: \n{b}\n')
```

```python
print(results.summary())
```

## How do we interpret the equation?

#### The R-squared in relatively large (R-squared=0.44).
#### The F-statistic is large (58.37) with p-value close to zero. Therefore the model is statistically significant.

#### R-squared is 0.44 (therefore 44% of the variance is explained by this model)

#### Education, experience and tenureship are statistically significant.

#### Female is not statistically significant (p-value>0.05). We observe than once we control for experience and tenureship, wages are not influenced by gender.

#### The interaction term female*education is not statistically significant (p-value>0.05). this means that the financial return of education is the same for males and females.

#### Both experience and tenureship exhibit quadratic behaviour (diminishing effects given that their coefficients are negative, -0.0006 and -0.0006 respectively).
